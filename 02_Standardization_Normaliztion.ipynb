## Standardiztion

### 01. Initial Phase

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

data = {
    "Age": [22, 25, 30, 35, 40, 45, 50, 60],
    "Salary": [25000, 27000, 30000, 32000, 35000, 37000, 40000, 100000],
    "Experience": [1, 2, 5, 7, 10, 12, 15, 20],
    "Department": ["HR", "IT", "IT", "HR", "Finance", "IT", "HR", "Finance"]
}

df = pd.DataFrame(data)
df.head(5)
```

### 02. Apply Standardization

### Approach - 1

```python
x = df.iloc[:, :-1]

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_dataset = pd.DataFrame(x_scaled, columns=x.columns)
x_dataset.head(5)
```

### 03. Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Before Standardization
plt.figure()
sns.kdeplot(data=df.iloc[:, :-1], fill=True)
plt.title('Salary Before Standardization')
plt.show()

# After Standardization
plt.figure()
sns.kdeplot(data=x_dataset, fill=True)
plt.title('Salary Before Standardization')
plt.show()
```
